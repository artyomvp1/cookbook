```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Click_income_pr2.xlsx')
df.head()
```

```python
plt.figure(figsize=(20,17))

#  
#plt.scatter(df['id'], df['amount'], color='Green', label='- ')
plt.plot(df['id'], df['amount'], color='Purple', label='- ')

#  
z = np.polyfit(df['id'], df['amount'], 2)
p = np.poly1d(z)
plt.plot(df['id'], p(df['id']), color='Red', linewidth=7, label='- ')

# 
plt.title('Misha', size=20, color='Red')
plt.xlabel('Date', size=20, color='Red')
plt.ylabel('Amount', size=20, color='Red')

plt.legend(loc='upper left')
plt.show()
```
