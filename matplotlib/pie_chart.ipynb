```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vgsales.csv')
df = df.groupby('Genre').sum('Global_Sales').reset_index(drop=False)
df
```

```python
# Parameters
x = [i for i in df['Global_Sales']]

labels = [i for i in df['Genre']]

autopct = '%.1f %%' # decimals

colors = ['yellowgreen', 'pink', 'gold', 'lightskyblue',
          'seagreen', 'lightcoral', 'lightblue', 'grey', 
          'purple', 'orange', 'plum', 'violet', 'magenta', 'cyan']

explode = [0.03 if df['Genre'][i] in ['Shooter', 'Sports'] else
           0.08 if df['Genre'][i] in ['Adventure', 'Puzzle'] else
           0 for i in range(len(df['Genre']))]

textprops = {'fontsize':15}

# Plot
plt.figure(figsize=(12, 12))
plt.pie(x, labels=labels, autopct=autopct, colors=colors, explode=explode, textprops=textprops)

# Donut setup
my_circle=plt.Circle((0,0), 0.45, color='white') # (0,0) - coordinated of the center, 0.45 - diameter
p=plt.gcf()
p.gca().add_artist(my_circle)

plt.show()
```
