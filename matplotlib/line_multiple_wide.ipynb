```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Line Graph

```python
gas = pd.read_csv('data_gas_price.csv')
gas.head(1)
```

```python
plt.figure(figsize=(13, 7))

for i in gas:
    if i != 'Year':
        plt.plot(gas['Year'], gas[i], '.-', label = i)
        
# Cosmetics
plt.title('Gas Price Over Time', fontdict={'fontweight':'bold', 'fontsize': 18})

plt.xlabel('YEAR')
plt.ylabel('US Dollars')

plt.xticks(gas.Year[::4].tolist() + [2020], rotation=45)

plt.legend()
plt.savefig('gas_price.png', dpi=300)
plt.show()
```
