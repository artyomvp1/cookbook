```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

# Categorical Plots
---
### 1. Bar Plot

```python
sns.barplot(data=tips_df, 
            x='sex', y='total_bill',
            estimator=np.mean)    # mean by default. np.std, np.median, np.variance
```

---
### 2. Count Plot

```python
sns.countplot(data=tips_df, 
              x='sex')    # Same as bar but estimator = count (occurances)
```

---
### 3. Box Plot

```python
sns.boxplot(data=tips_df, 
            x='day', y='total_bill', 
            hue='sex')    # division on male and female. Try without it

# Line in the box - median
# Box - standart deviation
# Vertical line - whiskers (data outside STD)
# Diamonds - outliers

plt.legend(loc=0)    # legend position
```

---
### 4. Violin Plot (Box Plot + KDE)

```python
sns.violinplot(data=tips_df, 
               x='day', y='total_bill',
               hue='sex')
```

```python
sns.violinplot(data=tips_df, 
               x='day', y='total_bill',
               hue='sex',
               split=True)    # combine what was divided by hue. To compare categories
```

---
### 5. Strip Plot (Scatter plot with one variable being categorical)

```python
plt.figure(figsize=(11,5))
sns.stripplot(data=tips_df, 
              x='day', y='total_bill',
              jitter=True)    # prevent dots overlaying a bit
```

```python
# With hue
plt.figure(figsize=(11,5))
sns.stripplot(data=tips_df, 
              x='day', y='total_bill',
              jitter=True,
              hue='sex')    # separete male and female
```

```python
# With hue on different plots
plt.figure(figsize=(11,5))
sns.stripplot(data=tips_df, 
              x='day', y='total_bill',
              jitter=True,
              hue='sex',
              dodge=True)    # separete male and female completely
```

---
### 6. Swarm Plot (Strip Plot + Violin)

```python
plt.figure(figsize=(8,5))
sns.swarmplot(data=tips_df,
              x='day', y='total_bill')
```

```python
# Stack swarm plot and violin plot
plt.figure(figsize=(8,5))
sns.swarmplot(data=tips_df,x='day', y='total_bill', color='white')
sns.violinplot(data=tips_df, x='day', y='total_bill')
```
