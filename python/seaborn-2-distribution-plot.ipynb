```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

# Distribution Plots
---
### 1. Regular Distribution Plot

```python
# Importing car crash dataset
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
# Histogram displot
sns.distplot(crash_df['not_distracted'], kde=True)    # (!)After typing distplot hit shift+Tab to get info about available attributes
```

The line is KDE (Kernel density estimation)

```python
sns.distplot(crash_df['not_distracted'], kde=False, bins=25)    # removing KDE, adding bins
```

---
### 2. Joint Plot

```python
# Regression
sns.jointplot(data=crash_df, 
              x='speeding', y='alcohol',    # axises
              kind='reg')    # reg

# Kernel density
sns.jointplot(data=crash_df, 
              x='speeding', y='alcohol',
              kind='kde')    # kde

# Hexogon distribution
sns.jointplot(data=crash_df, 
              x='speeding', y='alcohol',
              kind='hex')    # hex
```

---
### 3. KDE Plot

```python
sns.kdeplot(crash_df['alcohol'])
```

---
### 4. Pair Plot

```python
# Tips Dataframe
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
# Regular pairplot
sns.pairplot(tips_df)
```

```python
# Categorization using hue
sns.pairplot(tips_df, hue='sex', palette='inferno')    # HUE!!!
```

---
### 5. Rug Plot (Not popular)

```python
sns.rugplot(tips_df['tip'])
```
