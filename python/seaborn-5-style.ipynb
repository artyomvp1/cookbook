```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

# Styling
---

### 1. Grid Style

```python
plt.figure(figsize=(12,12))
sns.set_style('darkgrid')    # white, whitegrid, dark, darkgrid, ticks
sns.jointplot(data=crash_df, x='speeding', y='alcohol', kind='reg')
```

---
### 2. Font

```python
sns.set_context('paper', font_scale=1.9)    # Font size of axis
sns.jointplot(data=crash_df, x='speeding', y='alcohol', kind='reg')
```

```python
sns.set_context('talk', font_scale=1.9)    # talk, poster - eeasier for presentations
sns.jointplot(data=crash_df, x='speeding', y='alcohol', kind='reg')
```

---
### 3. Remove Axises

```python
sns.jointplot(data=crash_df, x='speeding', y='alcohol', kind='reg')

sns.despine(left=False)    # removes actual axis
```

---
### 4. Palettes

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
plt.figure(figsize=(8,5))
sns.set_style('darkgrid')
sns.set_context('talk')

sns.stripplot(data=tips_df, 
              x='day', y='total_bill', 
              hue='sex',
              palette='rocket')    # put palette color map
```
